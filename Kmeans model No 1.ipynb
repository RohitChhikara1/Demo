```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns # for statistical data visualization
%matplotlib inline
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples = 1000, centers = 3,n_features = 2, random_state = 23)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30, random_state = 23)
```

```python
from sklearn.cluster import KMeans
```

```python
## Manualy method
## Elbow method to find the value of K
## for avoiding warning we use this-
import warnings
warnings.filterwarnings('ignore')
```

```python
## wcss with in the cluster sum of square
wcss=[]
for k in range(1,11):
    Kmeans = KMeans(n_clusters = k,init = 'k-means++')
    Kmeans.fit(X_train)
    wcss.append(Kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlable('No of clusters')
plt.ylable('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3,init ='k-means++')
```

```python
y_train_lable = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c = y_train_lable)
```

```python
y_test_lable = kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c = y_test_lable)
```

```python
## Performance metrices
## Silhoutte score
from sklearn.metrics import silhouette_score
##from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
## find silhouette score
silhouette_score=[]
for k in range(1,11):
    Kmeans = KMeans(n_clusters = k,init = 'k-means++')
    Kmeans.fit(X_train)
    score = silhouette_score(X_train,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
## plotting silhoutte score
plt.plot(range(1,11),silhouette_coefficients)
plt.xticks(range(1,11))
plt.xlable('No of clusters')
plt.ylable('silhouette_coefficients')
plt.show()
```
